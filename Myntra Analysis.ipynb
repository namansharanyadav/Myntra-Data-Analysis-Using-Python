```python
orders["Month"]=orders["Date"].dt.strftime("%B")
orders
```

```python
gb=orders.groupby('Month').agg({"Original Price":"sum"})
gb
```

```python
plt.figure(figsize=(12,4))
sns.barplot(x=gb.index,y=gb['Original Price'], order=["January","February","March","April","May","June","July","August",'September',"October","November","December"],hue=gb.index,palette='Reds')
plt.xticks(rotation=30)
plt.show()
```
